**Approach of imputing missing values**

When addressing missing values, the imputation strategy varies for numerical and categorical columns are different.For numerical coumns, utilizing the median or mean for imputation is common. Imputing the median values is considered a superior approch compared to using the mean, as the median is less sensitive to outliers, resulting in a more robust imputation method.For categorical columns, mode is the best method for imputation

**Calculating Wins points**

```python
#creating function to calculate points by scores in the game


def calculate_points(row):
    if row['Home_WP_post'] > row['Away_WP_post']:
        return 2,0
    elif row['Home_WP_post']< row['Away_WP_post']:
        return 0,2
    else:
        if row['PosTeamScore'] > row['DefTeamScore']:
           return 2,0
        elif row['PosTeamScore'] > row['DefTeamScore']:
           return 2,0
        else:
              return 1,1

final_scores['home_team_points'], final_scores['away_team_points']=zip(*final_scores.apply(calculate_points,axis=1))
```

```python
filtered_games=final_scores[(final_scores['home_team_points']==1)| (final_scores['away_team_points']==1)]
filtered_games.head(15)
```

**Which team is best?**

```python
#create a new Dataframe with the total points for both have home and away teams
home_points = final_scores[['HomeTeam','home_team_points']].rename(columns={'HomeTeam':'Team','home_team_points':'Points'})
Away_points = final_scores[['AwayTeam','away_team_points']].rename(columns={'Away_points':'Team','away_team_points':'Points'})

#concatenate the home and away points Dataframe
all_points=pd.concat([home_points,Away_points])

#Group by team and sum the points
total_points_df=all_points.groupby('Team').agg(Total_points=('Points','sum')).reset_index()

#sort by total points in descending order
total_points_df = total_points_df.sort_values(by='Total_points',ascending=False)

print(total_points_df)
```

```python
#suppress warnings
import warnings
warnings.filterwarnings("ignore")

#plotting with seaborn
plt.figure(figsize=(12,8))
sns.set(style="whitegrid")
plt.xticks(rotation=90)

#create aa bar plot
bar_plot= sns.barplot(
    x='Team',
    y='Total_points',
    data=total_points_df,
    palette=sns.color_palette("RdYlGn",n_colors=len(total_points_df))
)

#Add Labels and title
bar_plot.set_xlabel('Team')
bar_plot.set_ylabel('Total Points')
bar_plot.set_title('Total Points Scored by Team 2009-2007')
```

**Win Percentage Comparison Between Home3 and Away Games**


```python
#Calculate home and away wins
home_wins=final_scores[final_scores['home_team_points'] > final_scores['away_team_points']].shape[0]
away_wins=final_scores[final_scores['away_team_points'] > final_scores['home_team_points']].shape[0]


#calculate total games
total_games =len(final_scores)

#calculate win percentage
win_percentages = {
    'Home': (home_wins/total_games)*100,
    'Away': (away_wins/total_games)*100,
}

#create Dataframe from the dictionary
win_data = pd.DataFrame(list(win_percentages.items()),columns=['Location','Win Percentage'])

#plotting with seaborn
plt.figure(figsize=(8,6))
sns.set(style="whitegrid")

#create a bar plot
bar_plot = sns.barplot(
    x='Location',
    y='Win Percentage',
    data=win_data,
    palette=['Green','Orange']
)

#Add Labels and title
bar_plot.set_xlabel('Location')
bar_plot.set_ylabel('Win Percentage')
bar_plot.set_title('Win Percentage Comparison: Home VS Away')
bar_plot.set(ylim=(0,100))



```

Insights:

The plot does suggest that the home team generally has a higher win percentage than the away team. This is a well-known phenonmenon in American Football,
sometimes referred to as "home field advantage".This might be happen because of-

  1.The home team gets to play in their own stadium,where thet are familliar with the field conditions, crowd noise, and other factors.
  2.The away team has to travel has to travel to the game, which can be tiring and disruptive to their routine.
  3.Home teams typically have more fan support than away team, which can create a more positive and energizing atmosphere.
